### Dataset Attributes

Understanding each column helps you make better decisions during cleaning and feature engineering.

The Wine dataset contains 13 chemical analysis measurements of wines from 3 different cultivars in Italy:

| # | Attribute | Description |
|---|-----------|-------------|
| 1 | **alcohol** | Alcohol content (% by volume). Typically ranges 11–15%. |
| 2 | **malic_acid** | Malic acid (g/L). An organic acid found in grapes; higher in cooler climates. |
| 3 | **ash** | Ash content (g/L). Inorganic residue after combustion; reflects mineral content. |
| 4 | **alcalinity_of_ash** | Alcalinity of ash (mEq/L). Measures the alkaline earth metals in ash. |
| 5 | **magnesium** | Magnesium (mg/L). An essential mineral; varies by soil and cultivar. |
| 6 | **total_phenols** | Total phenols. Antioxidant compounds; contribute to color, flavor, and mouthfeel. |
| 7 | **flavanoids** | Flavanoids. A subclass of phenols; major contributor to red wine color and taste. |
| 8 | **nonflavanoid_phenols** | Nonflavanoid phenols. Phenolic compounds that are not flavanoids. |
| 9 | **proanthocyanins** | Proanthocyanins. Tannin-like compounds; affect astringency. |
| 10 | **color_intensity** | Color intensity. Optical density measurement; indicates wine richness. |
| 11 | **hue** | Hue. Ratio of optical densities at different wavelengths; indicates wine age/type. |
| 12 | **od280_od315** | OD280/OD315 of diluted wines. Protein content indicator via UV absorption ratio. |
| 13 | **proline** | Proline (mg/L). An amino acid; higher in wines from warmer/drier regions. |

- **class** (target): Wine cultivar — 1, 2, or 3 (encoded as 0, 1, 2 for modeling)
  - Class 1 (cultivar_1): 59 samples
  - Class 2 (cultivar_2): 71 samples  
  - Class 3 (cultivar_3): 48 samples
