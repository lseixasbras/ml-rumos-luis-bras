### Dataset Attributes

Understanding each column helps you make better decisions during cleaning and feature engineering.

The Wine dataset contains 13 chemical analysis measurements of wines from 3 different cultivars in Italy:

| # | Attribute | Description |
|---|-----------|-------------|
| 1 | **alcohol** | Alcohol content (% by volume). Typically ranges 11–15%. |
| 2 | **malic_acid** | Malic acid (g/L). An organic acid found in grapes; higher in cooler climates. |
| 3 | **ash** | Ash content (g/L). Inorganic residue after combustion; reflects mineral content. |
| 4 | **alcalinity_of_ash** | Alcalinity of ash (mEq/L). Measures the alkaline earth metals in ash. |
| 5 | **magnesium** | Magnesium (mg/L). An essential mineral; varies by soil and cultivar. |
| 6 | **total_phenols** | Total phenols. Antioxidant compounds; contribute to color, flavor, and mouthfeel. |
| 7 | **flavanoids** | Flavanoids. A subclass of phenols; major contributor to red wine color and taste. |
| 8 | **nonflavanoid_phenols** | Nonflavanoid phenols. Phenolic compounds that are not flavanoids. |
| 9 | **proanthocyanins** | Proanthocyanins. Tannin-like compounds; affect astringency. |
| 10 | **color_intensity** | Color intensity. Optical density measurement; indicates wine richness. |
| 11 | **hue** | Hue. Ratio of optical densities at different wavelengths; indicates wine age/type. |
| 12 | **od280_od315** | OD280/OD315 of diluted wines. Protein content indicator via UV absorption ratio. |
| 13 | **proline** | Proline (mg/L). An amino acid; higher in wines from warmer/drier regions. |

- **class** (target): Wine cultivar — 1, 2, or 3 (encoded as 0, 1, 2 for modeling)
  - Class 1 (cultivar_1): 59 samples
  - Class 2 (cultivar_2): 71 samples  
  - Class 3 (cultivar_3): 48 samples



```python
# TODO: Load the raw CSV data into a DataFrame
# The wine.data file has no header row, so we pass header=None and assign column names ourselves.
# The first column is the class identifier (1-3), followed by 13 chemical measurements.

RAW_DIR = Path("../data/raw")

attribute_names = [
    "class", "alcohol", "malic_acid", "ash", "alcalinity_of_ash",
    "magnesium", "total_phenols", "flavanoids", "nonflavanoid_phenols",
    "proanthocyanins", "color_intensity", "hue", "od280_od315", "proline"
]

# Load the data with pd.read_csv()
# Use header=None and names=attribute_names.
# After loading, check the shape and preview the first few rows.
# Also run .info() to confirm all numeric columns have appropriate dtypes.

df = pd.read_csv(RAW_DIR / "wine.data", header=None, names=attribute_names)
print(f"Shape: {df.shape}")
print(df.head())
df.info()

```

```python
# TODO: Encode the target variable as integers (0, 1, 2)
# The class column has three unique values: 1, 2, 3 (wine cultivars).
# We map these to 0, 1, 2 so models can process them with zero-indexed classes.
# Label encoding (0/1/2) is fine here because the classes have no ordinal relationship
# — any order works for a 3-class classifier that uses softmax.

# Example encoding logic:
# class_mapping = {1: 0, 2: 1, 3: 2}
# df["class"] = df["class"].map(class_mapping)

# TODO: Check the class distribution
# The dataset has 59/71/48 samples per class — verify this stayed correct after encoding.

```

```python
# Encode target: map class 1,2,3 → 0,1,2
class_mapping = {1: 0, 2: 1, 3: 2}
df["class"] = df["class"].map(class_mapping)

# Verify encoding
print("Unique classes after encoding:", df["class"].unique())
print("\nClass distribution:")
print(df["class"].value_counts().sort_index())
```

```python
# TODO: Use .value_counts() to check the distribution of the encoded classes
```

```python
# Check class distribution with value_counts
print("Class distribution (value_counts):")
print(df["class"].value_counts().sort_index())
print(f"\nTotal samples: {len(df)}")
print(f"Class balance: {df['class'].value_counts(normalize=True).sort_index().round(3).to_dict()}")
```

```python
# TODO: Check for missing values in each column
# Use df.isnull().sum() to get a count of NaN entries per column.
# The original wine.data has no missing values, but it's good practice to verify.


# TODO: Handle missing values if any are found
# Two common strategies:
#   - df.dropna() — removes any row with a missing value (simple, but you lose data)
#   - df.fillna(value) — replaces missing entries with something like the column mean or median
# Choose the approach that makes sense for your data.

# TODO: Verify the final data is clean using df.info() and df.describe()
```

```python
# Check for missing values
print("Missing values per column:")
missing = df.isnull().sum()
print(missing)
print(f"\nTotal missing values: {missing.sum()}")

# If any missing values exist, handle them
if missing.sum() > 0:
    # Strategy: fill numeric columns with median (robust to outliers)
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
    print("\nMissing values filled with column medians.")
else:
    print("\nNo missing values found — data is complete.")

# Verify data is clean
print("\n--- Data Summary ---")
df.info()
print("\n--- Descriptive Statistics ---")
df.describe()
```

```python
# TODO: Find duplicate rows in the DataFrame
# Use df.duplicated() to identify duplicate rows. 
# This returns a boolean Series where True indicates a duplicate row (except for the first occurrence).
# You can sum this Series to get the total count of duplicates.


# TODO: Duplicates can be searched for across all columns or a subset of columns.
# e.g. df.duplicated(subset=["Date", "Hour"])

# TODO: Remove duplicate rows if any are found
# Use df.drop_duplicates() to remove duplicate rows.
# param: keep="first": it keeps the first occurrence and drops subsequent duplicates. alternatives are "last" (keeps the last occurrence) or False (drops all duplicates).
# You can specify subset=["Date", "Hour"] to drop duplicates based on just those columns.

# TODO: Alternatively, you can use df.duplicated() to filter the DataFrame and inspect duplicates before dropping them.

# TODO: Reset the index after dropping duplicates
```

```python
# Find duplicate rows
duplicates = df.duplicated()
num_duplicates = duplicates.sum()
print(f"Number of duplicate rows: {num_duplicates}")

if num_duplicates > 0:
    # Inspect duplicates before dropping
    print("\nDuplicate rows:")
    print(df[duplicates])
    
    # Remove duplicates, keep the first occurrence
    df = df.drop_duplicates(keep="first")
    print(f"\nAfter removing duplicates: {df.shape[0]} rows remaining")
    
    # Reset index
    df = df.reset_index(drop=True)
    print("Index reset.")
else:
    print("No duplicates found — all rows are unique.")

print(f"\nFinal shape: {df.shape}")
```

It is often spread across multiple tables in a database — a **fact** table 
and several **dimension** tables, linked by key columns.

To build your feature matrix you need to **join** these tables back together.

In this section you will work with a relational version of the Wine data,
split into multiple CSV files that simulate a database:

| File | Content | Key column |

| `cultivar.csv` | Cultivar class label | `sample_id` |

```python
# Load relational tables
RELATIONAL_DIR = Path("../data/raw/relational")

cultivar = pd.read_csv(RELATIONAL_DIR / "cultivar.csv")
chemical = pd.read_csv(RELATIONAL_DIR / "chemical.csv")
phenols = pd.read_csv(RELATIONAL_DIR / "phenols.csv")

# Inspect each table
print(f"Cultivar: {cultivar.shape}")
print(cultivar.head())
print(f"\nChemical: {chemical.shape}")
print(chemical.columns.tolist())
print(chemical.head())
print(f"\nPhenols: {phenols.shape}")
print(phenols.columns.tolist())
print(phenols.head())
```

```python
# TODO: Merge cultivar with chemical measurements
# Use pd.merge() with left_on="sample_id" and right_on="chemical_id".
# Use a LEFT JOIN to keep all specimens.
# Count how many rows have NaN for chemical columns.

# merged = pd.merge(cultivar, chemical, left_on="sample_id", right_on="chemical_id", how="left")

# TODO: Now try an INNER JOIN. How many rows do you lose? Why?

```

```python
# LEFT JOIN: Merge cultivar with chemical measurements
merged = pd.merge(cultivar, chemical, left_on="sample_id", right_on="chemical_id", how="left")
print(f"After LEFT JOIN (cultivar + chemical): {merged.shape}")

# Count rows with missing chemical data
chemical_cols = ["alcohol", "malic_acid", "ash", "alcalinity_of_ash", "magnesium", "color_intensity", "proline"]
missing_chemical = merged[chemical_cols].isna().any(axis=1).sum()
print(f"Rows missing chemical data: {missing_chemical}")

# INNER JOIN comparison
merged_inner = pd.merge(cultivar, chemical, left_on="sample_id", right_on="chemical_id", how="inner")
print(f"\nAfter INNER JOIN (cultivar + chemical): {merged_inner.shape}")
print(f"Rows lost with INNER JOIN: {len(cultivar) - len(merged_inner)}")
```

```python
# TODO: Merge in the phenol measurements
# Chain a second merge to bring in phenol data.

# TODO: Check how many rows are missing phenol data
```

```python
# Merge in phenol measurements (chain second merge)
merged = pd.merge(merged, phenols, left_on="sample_id", right_on="phenol_id", how="left")
print(f"After LEFT JOIN (+ phenols): {merged.shape}")

# Count rows missing phenol data
phenol_cols = ["total_phenols", "flavanoids", "nonflavanoid_phenols", "proanthocyanins", "hue", "od280_od315"]
missing_phenols = merged[phenol_cols].isna().any(axis=1).sum()
print(f"Rows missing phenol data: {missing_phenols}")

# Show columns in final merged table
print(f"\nAll columns: {merged.columns.tolist()}")
print(f"\nFirst few rows:")
merged.head()
```

```python
# TODO: Compare to the original single-table shape
# The original wine.data has 178 rows with 13 features + class.
# After splitting and rejoining, how many complete rows do you have?

# HINT: This tells you how many specimens have all data sources.

```

```python
# Compare to the original single-table shape
print(f"Original wine.data: {df.shape} (178 rows × 14 columns incl. class)")
print(f"Merged relational:  {merged.shape}")

# Count complete rows (no NaN in any feature column)
all_feature_cols = chemical_cols + phenol_cols
complete_rows = merged.dropna(subset=all_feature_cols)
print(f"\nComplete rows (all features present): {len(complete_rows)} / {len(merged)}")
print(f"Rows with at least one missing feature: {len(merged) - len(complete_rows)}")

# Check if any row is missing BOTH chemical and phenol data
both_missing = (merged[chemical_cols].isna().any(axis=1) & merged[phenol_cols].isna().any(axis=1)).sum()
print(f"Rows missing data from BOTH tables: {both_missing}")
```
